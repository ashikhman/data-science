```python
import numpy as np

weight = np.array([0.5])
input = np.array([0.8])

pred = np.array([0.4])
goal = np.array([1.0])

def neural_network(input, weight):
    pred = input * weight
    return pred

lr = 0.01
for i in range(76):
    pred = neural_network(input, weight)
    error = (pred - goal) ** 2
    delta = pred - goal

    print(i, "Error:", error, "Pred:", pred)

    pred_up = neural_network(input, weight + lr)
    pred_down = neural_network(input, weight - lr)

    for i in range(len(pred)):
        error = error[i]
        error_up = (pred_up[i] - goal[i]) ** 2
        error_down = (pred_down[i] - goal[i]) ** 2
        delta_up = pred_up[i] - goal[i]
        delta_down = pred_down[i] - goal[i]

        if error > error_up or error > error_down:
            if error_up < error_down:
                weight[i] += lr
            else:
                weight[i] -= lr

print(weight)
```
